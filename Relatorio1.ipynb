```
#Métodos Aqui!


#************************************MÉTODOS DE ITERAÇÃO******************************************

def bisectionIterationMethod(a, b, f, numIter):
 
  printTitle("\t\t\t************************MÉTODO DA BISSECÇÃO********************************\n")

  for i in range(numIter):
    x = (a + b)/2
    fx = f(x)
    fa = f(a)
    fafx = fa*fx

    printResults("i=%3d ->  a:%10.6f | b:%10.6f | x: %10.6f | f(x): %10.6f | f(a):%10.6f | f(a)*f(x):%10.6f" % (i, a, b, x, fx, fa, fafx))

    if(fafx < 0):
      b = x
    else:
      a = x

  printTitle("\n\t\t\t***************************************************************************\n\n")



def falsePositionIterationMethod(a, b, f, numIter):

  printTitle("\t\t\t************************MÉTODO DA FALSA POSIÇÃO****************************\n")

  for i in range(numIter):
    fa = f(a)
    fb = f(b)
    x = ((a * fb) - (b*fa))/(fb - fa)
    fx = f(x)
    fafx = fa*fx

    printResults("i=%3d ->  a:%10.6f | b:%10.6f | f(a):%10.6f | f(b):%10.6f | x: %10.6f | f(x): %10.6f | f(a)*f(x):%10.6f" %
          (i, a, b, fa, fb, x, fx, fafx))

    if(fafx < 0):
      b = x
    else:
      a = x

  printTitle("\n\t\t\t***************************************************************************\n\n")



def tangentIterationMethod(x0, f, numIter):

    printTitle("\t\t\t************************MÉTODO DA TANGENTE*******************************\n")

    for i in range(numIter):

      x = sympy.symbols('x')
      fx = f(x0)
      dfx = sympy.diff(f(x), x).subs(x, x0)
      x1 = x0 - fx/dfx

      printResults("i=%3d ->  x0:%10.6f | x1:%10.6f | f(x): %10.6f | f(x)':%10.6f" % (i, x0, x1, fx, dfx))

      x0 = x1


    printTitle("\n\t\t\t*************************************************************************\n\n")



def secantIterationMethod(x0, x1, f, numIter):

  printTitle("\t\t\t************************MÉTODO DA SECANTE**********************************\n")

  fx0 = f(x0)
  fx1 = f(x1)

  for i in range(numIter):

      x2 = x1 - (fx1*(x1 - x0))/(fx1-fx0)
      fx2 = f(x2)

      printResults("i=%3d ->  x0:%10.6f | x1:%10.6f | x2:%10.6f | f(x0): %10.6f | f(x1):%10.6f | f(x2):%10.6f" % (i, x0, x1, x2, fx0, fx1, fx2))

      x0 = x1
      x1 = x2
      fx0 = fx1
      fx1 = fx2

  printTitle("\n\t\t\t***************************************************************************\n\n")





#************************************MÉTODOS DE PRECISÃO******************************************


def bisectionPrecisionMethod(a, b, f, precision):
 
  printTitle("\t\t\t************************MÉTODO DA BISSECÇÃO********************************\n")

  i = 0
  fx = sys.float_info.max

  while(math.fabs(fx) > precision):
    x = (a + b)/2
    fx = f(x)
    fa = f(a)
    fafx = fa*fx

    printResults("i=%3d ->  a:%10.6f | b:%10.6f | x: %10.6f | f(x): %10.6f | f(a):%10.6f | f(a)*f(x):%10.6f" % (i, a, b, x, fx, fa, fafx))

    if(fafx < 0):
      b = x
    else:
      a = x

    i+=1

  printTitle("\n\t\t\t***************************************************************************\n\n")

  return i;



def falsePositionPrecisionMethod(a, b, f, precision):

  printTitle("\t\t\t************************MÉTODO DA FALSA POSIÇÃO****************************\n")

  i = 0
  fx = sys.float_info.max

  while(math.fabs(fx) > precision):
    fa = f(a)
    fb = f(b)
    x = ((a * fb) - (b*fa))/(fb - fa)
    fx = f(x)
    fafx = fa*fx

    printResults("i=%3d ->  a:%10.6f | b:%10.6f | f(a):%10.6f | f(b):%10.6f | x: %10.6f | f(x): %10.6f | f(a)*f(x):%10.6f" %
          (i, a, b, fa, fb, x, fx, fafx))

    if(fafx < 0):
      b = x
    else:
      a = x
    
    i+=1

  printTitle("\n\t\t\t***************************************************************************\n\n")

  return i;



def tangentPrecisionMethod(x0, f, precision):

  printTitle("\t\t\t************************MÉTODO DA TANGENTE*********************************\n")

  fx = sys.float_info.max
  i = 0

  while(math.fabs(fx) > precision):

    x = sympy.symbols('x')
    fx = np.float32(f(x0))
    dfx = np.float32(sympy.diff(f(x), x).subs(x, x0))
    x1 = np.float32(x0 - fx/dfx)

    printResults("i=%3d ->  x0:%10.6f | x1:%10.6f | f(x): %18.6f | f(x)':%18.6f" % (i, x0, x1, fx, dfx))

    x0 = np.float32(x1)
    i+=1


  printTitle("\n\t\t\t***************************************************************************\n\n")

  return i;



def secantPrecisionMethod(x0, x1, f, precision):

  printTitle("\t\t\t************************MÉTODO DA SECANTE**********************************\n")

  fx0 = f(x0)
  fx1 = f(x1)

  fx2 = sys.float_info.max
  i = 0

  while(math.fabs(fx2) > precision):

      x2 = x1 - (fx1*(x1 - x0))/(fx1-fx0)
      fx2 = f(x2)

      printResults("i=%3d ->  x0:%10.6f | x1:%10.6f | x2:%10.6f | f(x0): %10.6f | f(x1):%10.6f | f(x2):%10.6f" % (i, x0, x1, x2, fx0, fx1, fx2))

      x0 = x1
      x1 = x2
      fx0 = fx1
      fx1 = fx2

      i+=1


  printTitle("\n\t\t\t***************************************************************************\n\n")

  return i;

def callAllPrecisionMethods(a, b, f, f_sympy, precision):
  iterationsOnMethod = []

  iterationsOnMethod.append(bisectionPrecisionMethod(a, b, f, precision))
  iterationsOnMethod.append(falsePositionPrecisionMethod(a, b, f, precision))
  iterationsOnMethod.append(tangentPrecisionMethod(a, f_sympy, precision))
  iterationsOnMethod.append(secantPrecisionMethod(a, b, f, precision))

  return iterationsOnMethod

def graphicalPlot(f, start, stop, numPoints):
  curve = np.linspace(start, stop, numPoints)
  zeroes = []
  zeroesPos = []
  values = []
  countPoints = 0

  for value in curve:
    fValue = f(value)
    #print(fValue)
    if(fValue >= -0.001 and fValue <= 0.001):
      zeroes.append(value)
      zeroesPos.append(fValue)
    values.append(fValue)
    countPoints += 1

  plot.plot(curve, values, linestyle="-")
  plot.plot(zeroes, zeroesPos,  linestyle='none', marker='s', color = 'black')
  for x,y in zip(zeroes, zeroesPos): 
    plot.text(x, y, "%.6f"%(x), color='green')
  plot.title("Método gráfico", color='yellow', fontweight='bold')
  plot.grid(True)
  plot.tick_params(axis='x', colors='yellow')
  plot.tick_params(axis='y', colors='yellow')
  plot.xlabel('x', color='green', fontweight='bold')
  plot.ylabel('f(x)', color='green', fontweight='bold')
  plot.show()

def printStatistics(methodsListOfLists):

  print(colored("\n\nDADOS ESTATÍSTICOS DO NÚMERO DE ITERAÇÕES PARA CADA MÉTODO NAS EQUAÇÕES DA QUESTÃO 5\n", 'white', attrs=['bold']))

  fieldsStr = "\t Método   |        Média |    Mediana |  Variância | Desvio Padrão"
  print(colored(fieldsStr, 'yellow', attrs=['bold']))

  for i in methodsListOfLists:
    statisticStr = "%15s   |   %10.3f | %10.3f | %10.3f | %10.3f" % (i[4], i[0], i[1], i[2], i[3])
    print(colored(statisticStr, 'green', attrs=['bold']))

def calcMean(results):
  sumAll = 0
  for i in results:
    sumAll += i
  mean = sumAll/results.__len__()
  return mean

def calcMedian(results):
  sortedResults = sorted(results)
  pos1 = int((results.__len__())/2)
  pos2 = int((results.__len__())/2) + 1
  median = (sortedResults[pos1] + sortedResults[pos2])/2
  return median

def calcVariance(results, mean):
  numeratorVar = 0
  for i in results:
    numeratorVar += math.pow(i - mean, 2)
  variance = numeratorVar/(results.__len__())
  return variance

def calcStdDeviation(variance):
  stdDeviation = math.sqrt(variance)
  return stdDeviation

def printTitle(title):
  print(colored(title, 'yellow', attrs=['bold'])) 

def printResults(resultStr):
  print(colored(resultStr, 'green'))

def printEquation(equation):
  print(colored(equation, 'cyan', attrs=['bold']))
```

```
f1 = lambda x: x**10 -1

bisectionIterationMethod(0, 1.3, f1, 5);
falsePositionIterationMethod(0, 1.3, f1, 5);
```

## Questão 5 – Compare o desempenho dos métodos da Bissecção, Falsa Posição, Tangente e Secante nas equações abaixo:
1. f(x)=e^(-x^2)-cos(x) 𝜀1=𝜀2=10^-4 𝜉 ∈ [1, 2]
2. f(x)=x^3-x-1         𝜀1=𝜀2=10^-6 𝜉 ∈ [1, 2] 
3. f(x)=4sen(x)-e^x     𝜀1=𝜀2=10^-5 𝜉 ∈ [0, 1] 
4. f(x)=xlog(x)-1       𝜀1=𝜀2=10^-7 𝜉 ∈ [2, 3]

 

Média dos tempos de execução dos métodos:


```
iterationsOnMethodsAndEquations = []

f5_1 = lambda x: math.exp(-x**2) - math.cos(x)
f5_1_sympy = lambda x: sympy.exp(-x**2) - sympy.cos(x)
printEquation("1.f(x)=e^(-x^2)-cos(x) 𝜀1=𝜀2=10^-4 𝜉 ∈ [1, 2]\n")
iterationsOnMethodsAndEquations.append(callAllPrecisionMethods(1, 2, f5_1, f5_1_sympy, 0.0001))

f5_2 = f5_2_sympy = lambda x: x**3 - x - 1 
printEquation("\n2.f(x)=x^3-x-1 𝜀1=𝜀2=10^-6 𝜉 ∈ [1, 2]\n")
iterationsOnMethodsAndEquations.append(callAllPrecisionMethods(1, 2, f5_2, f5_2_sympy, 0.000001))

f5_3 = lambda x: 4*math.sin(x) - math.exp(x)
f5_3_sympy = lambda x: 4*sympy.sin(x) - sympy.exp(x)  
printEquation("\n3.f(x)=4sen(x)-e^x 𝜀1=𝜀2=10^-5 𝜉 ∈ [0, 1]\n")
iterationsOnMethodsAndEquations.append(callAllPrecisionMethods(0, 1, f5_3, f5_3_sympy, 0.00001))

f5_4 = lambda x: x*math.log(x,10) - 1
f5_4_sympy = lambda x: x*sympy.log(x, 10) - 1
printEquation("\n4.f(x)=xlog(x)-1 𝜀1=𝜀2=10^-7 𝜉 ∈ [2, 3]")
iterationsOnMethodsAndEquations.append(callAllPrecisionMethods(2, 3, f5_4, f5_4_sympy, 0.0000001))

bisectionIterations = []
falsePositionIterations = []
tangentIterations = []
secantIterations = []
methodsIterations = []

for i in iterationsOnMethodsAndEquations:
  bisectionIterations.append(int(i[0]))
  falsePositionIterations.append(int(i[1]))
  tangentIterations.append(int(i[2]))
  secantIterations.append(int(i[3]))

bisectionIterations.append("Bissecção")
falsePositionIterations.append("Falsa Posição")
tangentIterations.append("Tangente")
secantIterations.append("Secante")

'''
print(bisectionIterations)
print(falsePositionIterations)
print(tangentIterations)
print(secantIterations)
'''
methodsIterations.append(bisectionIterations)
methodsIterations.append(falsePositionIterations)
methodsIterations.append(tangentIterations)
methodsIterations.append(secantIterations)

methodsListOfLists = []

for i in methodsIterations:
  methodStats = []
  mean = calcMean(i[0:4])
  methodStats.append(mean)
  methodStats.append(calcMedian(i[0:4]))
  variance = calcVariance(i[0:4], mean)
  methodStats.append(variance)
  methodStats.append(calcStdDeviation(variance))
  methodStats.append(i[4])
  methodsListOfLists.append(methodStats)

printStatistics(methodsListOfLists)
```

A tabela imediatamente acima mostra os resultados obtidos da aplicação de algumas métricas estatísticas sobre o número de iterações utilizados para cada caso de equação demonstrado. É bom salientar, ainda, que pelo fato da precisão mudar de equação para equação, isso pode influenciar nos resultados obtidos, não sendo a precisão igual para todas as equações a questão 5.
Dito isso, a tabela com os dados estatísticos nos mostra uma clara tendência decrescente nos valores de todas as métricas quando observamos do primeiro método implementado(Bissecção) ao último(Secante), tendo o método da Secante os melhores(e menores) valores de média, mediana, variância e desvio padrão para as equações utilizadas na questão 5.  


Um adendo: no método da tangente foi utilizada a biblioteca SymPy para calcular a derivada, por isso que também foram utilizados os métodos exp, sin, cos e log da SymPy para as funções calculadas pelo método da tangente.
